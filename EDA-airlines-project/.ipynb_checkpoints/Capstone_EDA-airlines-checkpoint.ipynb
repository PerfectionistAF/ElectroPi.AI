```python
#if csv/if excel/if sql/ allow for reading different types of files
def read (filename, filetype):
    if filetype == '.csv':
        df = pd.read_csv(filename)
    elif filetype == '.xslx':
        df = pd.read_excel(filename)
    elif filetype == '.sql':
        #df = pd.read_sql(sql, con, index_col=None) ##get the right libraries
        print('Please use sql engine and connect to sql database')
    return df
```

### Describe and identify types of columns:

```python
#return basic values and calculations
def describe(Dataframe):
    #check it is a dataframe
    pd.DataFrame(Dataframe)
    summary = Dataframe.describe()
    return summary
```

```python
#return list of column data types
def dtypes(Dataframe):
    pd.DataFrame(Dataframe)
    column_names = [Dataframe.dtypes]
    return column_names
```

```python
#return 2 lists, Numerical columns and Categorical columns
def types(Dataframe):
    pd.DataFrame(Dataframe)
    numerical = Dataframe.select_dtypes(exclude = ['object']).columns.tolist()
    categorical = Dataframe.select_dtypes(include = ['object']).columns.tolist()
    return numerical, categorical
```

### Handle nulls and duplicates 

```python
def checkduplicates(Dataframe):
    pd.DataFrame(Dataframe)
    duplicates = Dataframe.duplicated().tolist()
    if 'True' in duplicates:
        print('Duplicates found')
        indices = duplicates.index('True')
    else:
        print('No duplicates found')
        indices = 0
    return indices
```

```python
def checkna(Dataframe):
    pd.DataFrame(Dataframe)
    return Dataframe.isna().sum()
```

```python
#null values found in the carrier column: user can choose to either fill them manually
#or use median of numeric values and use max frequency for categorical values
###In case of PASSENGERS, null values will not be removed since carrier column has a 7% uniqueness
#so removing it will affect the values
def nullhandling(Dataframe):
    pd.DataFrame(Dataframe)
    manualhandling = input('Would you like to fill the null values in one value ? Y/y or N/n?' 
                             'If not an autofill with the median, for numerical features,' 
                             'and the maximum frequency, for categorical features, will be used.'
                          'Returns dataframe, n and c transforms, numerical and categorical lists of columns ')

    if manualhandling == 'y' or manualhandling == 'Y':
        fill = input('What value would you like to fill instead of the null values?')
        Full = Dataframe.fillna(fill)
        Numerical, Categorical = types(Dataframe)
        C = 0
        N = 0
    else:
        Full = Dataframe
        Numerical, Categorical = types(Dataframe)
        Imputer = SimpleImputer(missing_values = np.nan , strategy = 'median')
        N = Dataframe[Numerical] 
        Imputer.fit(N)
        N = Imputer.transform(N)
        Imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
        C = Dataframe[Categorical] 
        Imputer.fit(C)
        C = Imputer.transform(C)
    
    return Full, N, C, Numerical, Categorical
```

```python
#user selects feature as new index , sorts the data in that feature and prepares to visualise
#dimansionality reduction by LDA since data is scaled linearly, default by a user chosen feature
#user selects an index, gets the pearson correlation and removes features with a high correlation
#after that LDA is performed on the leftover table
def pearsonfeatureselection(TrainingSet, threshold):
    colcorr = []  
    corrmatrix = TrainingSet.corr()
    for i in range(len(corrmatrix.columns)):
        for j in range(i):
            if abs(corrmatrix.iloc[i, j]) > threshold: 
                colname = corrmatrix.columns[i]                  
                colcorr.append(colname)
        
    return colcorr
```

### Rename columns if needed

```python
#user can rename columns in dataframe  
def rename(Dataframe):
    pd.DataFrame(Dataframe)
    numerical, categorical = types(Dataframe)
    renamenumerical = input('Would you like to rename the numerical type columns, Y/y or N/n ?\n')
    if renamenumerical == 'Y' or renamenumerical == 'y':
        repeat = 'y'
        print('These are the numerical columns:\n', numerical)
        while repeat == 'y' or repeat == 'Y':
            column = input('Write the column here:\n ')
            newname = input('Write the new name here:\n ')
            if column in numerical:
                Dataframe = Dataframe.rename(columns = {column : newname})
                print('Successfully renamed')
                flag = 1
            else:
                flag = 0  ##unsuccessful renaming, try again
                print('Unsuccessfully renamed')
                
            repeat = input('Repeat?')
    renamecategorical = input('Would you like to rename the categorical type columns, Y/y or N/n ?\n')
    if renamecategorical == 'Y' or renamecategorical == 'y':
        repeat = 'y'
        print('These are the categorical columns:\n', categorical)
        while repeat == 'y' or repeat == 'Y':
            column = input('Write the column here:\n ')
            newname = input('Write the new name here:\n ')
            if column in categorical:
                Dataframe = Dataframe.rename(columns = {column : newname})
                print('Successfully renamed')
                flag = 1
            else:
                flag = 0  ##unsuccessful renaming, try again
                print('Unsuccessfully renamed')
                
            repeat = input('Repeat?')
    return Dataframe
```
